```python
import ast
```

```python
def modify(test):
    L=[]
    for i in ast.literal_eval(test):
        L.append(i["name"])
    return L
```

```python
df["genres"] = df["genres"].apply(modify)
df["keywords"] = df["keywords"].apply(modify)
```

```python
df.head()
```

```python
def edit(test):
    L=[]
    counter = 0
    for i in ast.literal_eval(test):
      if counter !=3:
         L.append(i["name"])
         counter +=1
      else:
        break
    return L
```

```python
df["cast"] = df["cast"].apply(edit)
```

```python
def clear(test):
    L=[]
    for i in ast.literal_eval(test):
        if i["job"]=="Director":
          L.append(i['name'])

    return L
```

```python
df["crew"] = df["crew"].apply(clear)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
df['genres'] = df['genres'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
```

```python
# Convert list elements to strings
df['genres'] = df['genres'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)

# Initialize CountVectorizer
cv = CountVectorizer()

# Fit and transform the data
vec = cv.fit_transform(df['genres'].values.astype('U')).toarray()

print(vec)
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vec)
```

```python
similarity
```

```python
df[df['title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vec:vec[1])
for i in distance[0:5]:
    print(df.iloc[i[0]].title)
```

```python
def recommend(movie_title):
    # Check if the movie exists in the DataFrame
    if movie_title not in df['title'].values:
        print(f"Movie '{movie_title}' not found in the database.")
        return

    # Find the index of the movie in the DataFrame
    index = df[df['title'] == movie_title].index[0]

    # Get the genre of the specified movie
    movie_genre = df.iloc[index]['genres']

    # Calculate the similarity distances
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vec: vec[1])

    # Print the title and genre of the specified movie
    print(f"Recommendations for '{movie_title}' Genre is: {movie_genre}")

    # Print the titles and genres of the top 5 similar movies
    for i in distance[1:6]:  # skipping the first result as it's the same movie
        title = df.iloc[i[0]]['title']
        genre = df.iloc[i[0]]['genres']
        print(f"Genre: {genre} , Title: {title}")

# Example usage
recommend("Spectre")
```

```python
# Saving the model
import pickle
with open('recmnd.pkl', 'wb') as file:
    pickle.dump(df, file)
with open('similarity.pkl', 'wb') as f:
    pickle.dump(similarity, f)
```
